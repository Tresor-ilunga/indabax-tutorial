## 2. Dimension  and type des données

```python
Ebola = pd.read_csv('data/ebola.csv')
```

Pour voir les n premières lignes d'un tableau il suffit d'utiliser la fonction **head()** et de specifier la valeur de n dans la parenthèse. Lorsqu'aucune valeur n'est specifier, la fonction *head()* affiche 5 premieres lignes. 

```python
Ebola.head(4) ##afficher 4 premieres lignes
```

Pour voir les n dernières lignes d'un tableau il suffit d'utiliser la fonction **tail()** et de specifier la valeur de n dans la parenthèse.

```python
Ebola.tail(6) ##afficher 6 dernieres lignes
```

Afficher les lignes aléatoirement on utilise la sonction **sample()**.

```python
Ebola.sample(3) ##afficher 3 lignes aléatoirement
```

Verifier la dimensin du tableau de données (numbre des lignes et colonnes)

```python
Ebola.shape
```

Verifier le type des donées de chaque colonnes, on utilise la fonction **info()**.

```python
Ebola.info()
```

* Nous pouvons constater ce tableau contient 122 entrées avec leur index allant de 0 à 121.

* Les colonnes **Cases_Guinea, Cases_Liberia, Cases_SierraLeone** and **Cases_Nigeria** sont de type **réel**.


* La colonne ***Day** est de type **entier** 

* Certaines colonnes ont des données manquantes:
        122-93 = 29 pour la colonne Cases_Guinea
        122-83 = 39 pour la colonne Cases_Liberia
        122-87 = 35 pour la colonne Cases_SierraLeone
        122-38 = 84 pour la colonne Cases_Nigeria

Verifier les noms de toutes les colonnes

```python
Ebola.columns
```

Afficher la moyenne, median, ecart type, valeur maximale, minimale, du tableau de données, on utilise la fonction **describe()*

```python
Ebola.describe()
```

Verifier les valeurs uniques d'une colonne.

```python
Ebola['Cases_Guinea'].unique()
```

Extraire une partie de données

```python
##Extraire le jour où le Guinné a eu un nombre de cas == 122
Ebola[Ebola['Cases_Guinea'] == 122]
```

```python
##Extraire les jour où Guiné a eu un nombre de cas > 200
Ebola[Ebola['Cases_Guinea'] > 200].head(10)
```

```python
##Extraire  les jours  où Guinné a eu un nombre de cas > 200 et Liberia >300
Ebola[(Ebola['Cases_Guinea'] >200)  & 
    ( Ebola['Cases_Liberia']> 300)]
```

A toi d'experimenter avec d'autres valeurs et aussi d'autres colonnes.


Supprimer une colonne

```python
Ebola = Ebola.drop('Cases_Nigeria' , axis = 1)  #axis=1 signifie colonne ( 0 pour ligne)
Ebola.head(5)
```

Créer une nouvelle colonne à partir des colonne existante

Pour notre tableaum, ajoutons 10 jours à chaque entrée de la colonne 'Day'

```python
Ebola['Day_10'] = Ebola['Day'] + 10
Ebola.head(5)
```
