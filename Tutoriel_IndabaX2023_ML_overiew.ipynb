## **Qu’est-ce que l'Intelligence Artificielle?**

L’Intelligence Artificielle (IA) est la simulation de processus de l’intelligence humaine par de machines. C’est programmer des machines à penser
comme des humains et imiter leurs actions.

Donc l’IA est l'ensemble des techniques qui imitent l'intelligence humaine et
permettent aux machines de reproduire des fonctions que l'on attribue aux
humains : voir, se déplacer, trier et hiérarchiser des informations, comprendre
un langage, prendre une décision.

## **Quelques exemples de la technologie de l’IA**

**L’automatisation:** qui fait qu’un système ou un processus fonctionne automatiquement.

**Le machine learning (Apprentissage automatique):** consiste à faire apprendre un ordinateur à partir de données sans le programmer.

**L'Apprentissage profondv( Deep Learning):** Un type particulier d’apprentissage automatique composé de réseaux neuronaux artificiels.

**La robotique:** la conception et la fabrication de robots. Qui vise à réaliser des agents qui peuvent agir comme des humains.

**La vision par ordinateur (Computer Vision):** capable de comprendre, interpréter et saisir les informations visuelles sur l’ordinateur.

**Le traitement automatique du langage
naturel (Natural Language Processing ou NLP):** est la capacité pour un programme informatique de comprendre le langage humain tel qu'il est parlé. Il est utilisé par exemple pour les traductions, la reconnaissance vocale ou encore les réponses automatiques aux questions.

##**Apprentissage Automatique (Machine Learning)**

L'apprentissage automatique est une branche de l'intelligence artificielle (IA) et de l'informatique qui porte sur l'utilisation de **données** et **d'algorithmes** pour imiter la manière dont les humains apprennent afin d'améliorer progressivement sa précision.

Les modèles d'apprentissage automatique appartiennent à quatre catégories principales.


1.   **Apprentissage automatique supervisé (Supervised Learning)**

L'apprentissage supervisé est l'utilisation d'ensembles de données étiquetées pour entraîner des algorithmes capables de classer des données ou de prédire des résultats avec précision.
![](https://drive.google.com/uc?export=view&id=1jEoVh6jtu_glsMoeB_A1IrThOBLSOjGD)



L’apprentissage supervisé est souvent utilisé pour deux types de problèmes: **les problèmes de régression** et **les problèmes de classification**.

* **Problème de régression:** lorsque la variable à expliquer $y$
est continue, $y ∈ R$. Par exemple lorsqu’on veut prédire le prix d’une bouteille de vin sur la base de certaines variables (le pays de fabrication, qualité, le taux d’alcool, etc).

* **Problème de classification:** A la différence avec le problème de régression, la classification est un autre type de problème
d’apprentissage supervisé où la variable à prédire est discrète (ou qualitative ou catégorique), $y \in \{ 0,1,2, ..\}$. Cette variable discrète peut être binaire (deux classes) ou multiple (multi-classe). Par exemple lorsqu’on veut
catégoriser si un e-mail reçu est un ’spam’ ou “non-spam” il s’agit bel et bien d’un problème de classification.

![](https://drive.google.com/uc?export=view&id=1sZNprk7f_WWD6IgZT6r07Mv_a4cRD3Ju)


2.  **Apprentissage automatique non-supervisé  (Unsupervised Learning)**

Est un type d’apprentissage automatique pour lequel les donneés
utilisées ne sont pas étiquetées. L’objectif de l’apprentissage est donc de découvrir des similitudesles et des structures sous-jacentes à ces données.
![](https://drive.google.com/uc?export=view&id=177KUxwdnxfe7PMOIj7QQd9y6ioiG6pH9)

**Exemple**

![](https://drive.google.com/uc?export=view&id=1TsYMcyAtIok4SV0Whm59cIQ5PYQS42d5)

L’algorithme de partitionnement comme le K-moyennes et les
algorithmes de réduction de dimension comme L'analyse en composantes principales (PCA, Principal Component Analysis) et la décomposition en valeurs singulières (SVD, Singular Value Cecomposition) sont les plus populaires.

3.   **Apprentissage semi-supervisé (Semi supervided Learning)**

L'apprentissage semi-supervisé est une combinaison de l'apprentissage supervisé et l'apprentissage non supervisé. Pendant l'entraînement, il utilise un jeu de données étiquetées plus petit pour guider la classification et l'extraction de caractéristiques à partir d'un jeu de données non étiquetées plus grand.

4.  **Apprentissage automatique par renforcement (Reinforcement Learning)**

Est un modèle d'apprentissage similaire à celui de l'apprentissage supervisé, mais l'algorithme n'est pas entraîné à l'aide d'échantillons de données. Ce modèle apprend au fur et à mesure de l'essai et l'erreur.

![](https://drive.google.com/uc?export=view&id=1VY-yHVI8ptEDDQVRc3GtdA_8UTu6Ktgj)





Signalons que **les réseaux de neurones**, **l'apprentissage en profondeur** et **l'apprentissage automatique** sont tous des sous-domaines de l'intelligence artificielle. Cependant, l'apprentissage en profondeur est un sous-domaine des réseaux de neurones et l'apprentissage automatique est un sous-domaine des réseaux de neurones.
